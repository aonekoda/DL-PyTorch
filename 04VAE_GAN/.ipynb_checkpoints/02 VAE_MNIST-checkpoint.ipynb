# VAE variational AutoEncoder

VAE는 Generative Model 중 하나로, 확률분포( P(x) )를 학습함으로써, 데이터를 생성하는 것이 목적이다.  
다음의 4개 코드 블럭이 사용된다.
1. data
2. model
3. training
4. inference

```python
import torch
from torchvision import datasets, transforms
import torchvision
import matplotlib.pylab as plt
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
```

## Data Load
데이터의 경우 MNIST 데이터 세트를 사용한다.

```python

# training data
BATCH_SIZE = 100

trainset = datasets.MNIST('./data/', train=True, download=True,
                   transform=transforms.ToTensor())
trainloader = torch.utils.data.DataLoader(trainset, batch_size=BATCH_SIZE,
                                          shuffle=True, num_workers=2)

# test data
testset = datasets.MNIST('./data/', train=False, download=False,
                   transform=transforms.ToTensor())

testloader = torch.utils.data.DataLoader(trainset, batch_size=BATCH_SIZE,
                                          shuffle=True, num_workers=2)
```

torchvision.utils.make_grid 여러 이미지에서 그리드를 만드는 방법 을 사용하여 데이터를 시각화한다.

```python
def show_images(images):
    images = torchvision.utils.make_grid(images)
    show_image(images[0])

def show_image(img):
    plt.imshow(img, cmap='gray')
    plt.show()
```

```python
dataiter = iter(trainloader)

images, labels = dataiter.next()

show_images(images)
```

## 모형의 생성
 VAE는 인코더와 디코더를 가지고 있다.
 
1. Encoder의 역할: 주어진 데이터를 잘 설명하는 잠재 변수의 분포를 찾는 것
2. Decoder의 역할: 두번째는 잠재변수로 부터 원본 이미지와 같은 이미지를 잘 복원하는 것이다.

---
## Reparameterization trick
가우시안 정규 분포의 샘플을 추출하고 싶을 때, 아래의 식과 같이 샘플링을 하는 것을 말한다. 이렇게 샘플을 추출 하더라도 원래의 확률적 특성을 보존 한다. 그리고 이렇게 샘플링 하면  z 는 확률분포의 모수인 분산과 평균이 더해진 형태이므로 Back propagation 또한 가능하다.

$$
sample\ z \tilde{} N(\mu, \sigma) <--> z = \mu + \epsilon\sigma $$
$$sample \ \epsilon \tilde{} \ N(0, 1)$$

```python
class VAE(nn.Module):
    def __init__(self, latent_variable_dim):
        super(VAE, self).__init__()
        self.fc1 = nn.Linear(784, 400)
        self.fc2m = nn.Linear(400, latent_variable_dim) # use for mean
        self.fc2s = nn.Linear(400, latent_variable_dim) # use for standard deviation
        
        self.fc3 = nn.Linear(latent_variable_dim, 400)
        self.fc4 = nn.Linear(400, 784)
        
    def reparameterize(self, log_var, mu):
        s = torch.exp(0.5*log_var)
        eps = torch.rand_like(s) # generate a iid standard normal same shape as s
        return eps.mul(s).add_(mu)
        
    def forward(self, input):
        x = input.view(-1, 784)
        x = torch.relu(self.fc1(x))
        log_s = self.fc2s(x)
        m = self.fc2m(x)
        z = self.reparameterize(log_s, m)
        
        x = self.decode(z)
        
        return x, m, log_s
    
    def decode(self, z):
        x = torch.relu(self.fc3(z))
        x = torch.sigmoid(self.fc4(x))
        return x
```

## Loss 함수

전체 네트워크가하는 일은 이미지를 인코딩 한 다음 코드를 사용하여 이미지를 재구성하는 것이다. 따라서 목표는 재구성을 원본 이미지에 최대한 가깝게 만드는 것이다. 손실 함수를 다음과 같이 구성한다.  

VAE는 두 가지 손실 유형을 합한 값이다.
1. 이미지 재구성 손실. 이것은 디코딩 된 이미지와 원본 이미지 사이의 단순한 크로스 엔트로피 (CE) 이다.
2. KL divergence:  latent variable z에 대한 것이다. 잠재 변수 z의 분포 P(z|x)가 정규분포와 유사한지 그 차이를 나타낸다.

```python
def loss(input_image, recon_image, mu, log_var):
    CE = F.binary_cross_entropy(recon_image, input_image.view(-1, 784), reduction='sum')
    KLD = -0.5 * torch.sum(1 + log_var - mu.pow(2) - log_var.exp())
    
    return KLD + CE
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
device
```

```python
vae = VAE(40)

vae
```

```python
train_on_gpu = torch.cuda.is_available()

if train_on_gpu:
    # move models to GPU
    vae.cuda()
    print('GPU available for training. Models moved to GPU')
else:
    print('Training on CPU.')
```

```python
learning_rate = 0.0001
optimizer = torch.optim.Adam(vae.parameters(), lr=learning_rate)
```

## Train Model
데이터를 읽고, 네트워크를 통해 이미지를 생성하고 loss를 계산하고 backpropagation을 수행한다. Adam 최적화를 사용한다.

```python
## train

train_loss = []
num_epochs = 5

for epoch in range(num_epochs):
    for i, data in enumerate(trainloader, 0):
        images, labels = data
        images = images.to(device)
        
        optimizer.zero_grad()
        
        recon_image, s, mu = vae(images)
        
        l = loss(images, recon_image, mu, s)
        
        l.backward()
        
        train_loss.append(l.item() / len(images))
        
        optimizer.step()
    print("Epoch [{:5d}/{:5d}] | loss: {:6.4f}".format(epoch+1, num_epochs, l.item()/len(images)))
```

```python
plt.plot(train_loss)
plt.show()
```

## Test the Model
몇 가지 재구성 된 이미지를 확인해 본다. 인코더와 디코더가 제대로 동작하여 이미지를 그럴듯하게 재구성하는지 출력한다.

```python
# Testing
with torch.no_grad():
    for i, data in enumerate(testloader, 0):
        images, labels = data
        images = images.to(device)
        recon_image, s, mu = vae(images)
        recon_image_ = recon_image.view(BATCH_SIZE, 1, 28, 28)
        if i % 100 == 0:
            show_images(recon_image_.cpu())
```

## 이미지를 재구성하기

디코더 만을 사용하여 임의의 이미지를 재구성해 본다. 랜덤하기 잠재 변수 z를 생성해서 decoder를 수행하면 이미지가 재구성 되는 것을 확인할 수 있다.

```python
# Just Generate
with torch.no_grad():
    z = np.random.randn(3,40)
    
    z = torch.FloatTensor(z)
    
    if train_on_gpu:
        z = z.cuda()
    
    sample_images = vae.decode(z)
    sample_images_ = sample_images.view(len(z), 1, 28,28)
    print(sample_images_.size())
    show_images(sample_images_.cpu())
```
